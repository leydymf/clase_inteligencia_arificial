```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# prompt: crear un dataframe df cargando

df = pd.read_csv('https://raw.githubusercontent.com/adiacla/bigdata/refs/heads/master/vivienda_clasificacion.csv')
df
```

```python
# prompt: crear una grafica de barras de habitaciones



# Count the occurrences of each room type
room_counts = df['habitaciones'].value_counts()

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(room_counts.index.astype(str), room_counts.values)
plt.xlabel('N√∫mero de Habitaciones')
plt.ylabel('Cantidad de Viviendas')
plt.title('Distribuci√≥n de Viviendas por N√∫mero de Habitaciones')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```
