# <font color="red">Cuaderno 1: Conceptos Básicos de EDA <p>

El análisis exploratorio de datos (EDA, por sus siglas en inglés) es un proceso crucial en el análisis de datos, ya que permite identificar las características principales de un conjunto de datos, visualizar patrones y realizar análisis estadísticos para una mejor comprensión. En este Sesion se cubren los conceptos básicos de EDA, incluyendo tipos de datos, visualización, medidas estadísticas y análisis de distribuciones.


## <font color="red">1.1 Identificación de tipos de datos <p>

![imagen](https://static.platzi.com/media/user_upload/Variables-e57fc8fe-6eb3-43e1-bf18-3362f3ed9da8.jpg)

En la grafica anterior se muestra como se clasifican los tipos de variables tanto cualitativa y cuantitativamente y como se pueden utilizar en los modelos de prediccion.

Ademas vamos a detallar como los datos se pueden clasificar en varias categorías según su naturaleza:


*   **Numéricos:** Representan valores cuantitativos.

Subcategorías:

**Continuos:** Pueden tomar cualquier valor en un rango (ej. altura, peso).

**Discretos:** Toman valores específicos (ej. número de hijos).


- **Categóricos nominales:**
Representan grupos o categorías. Ejemplo: género, color de ojos.

- **Categóricos Ordinales:**
Son categóricos, pero tienen un orden o jerarquía. Ejemplo: niveles de educación (primaria, secundaria, superior).

# <font color="red">1.2 Visualización de datos y sus patrones <p>
La visualización es fundamental en el EDA para detectar patrones, anomalías y relaciones entre variables. Algunas técnicas comunes incluyen:



* Histogramas: Útiles para visualizar la distribución de una variable numérica.
*   Gráficos de dispersión (scatter plots): Muestran relaciones entre dos variables numéricas.
* Diagramas de caja (boxplots): Ayudan a identificar valores atípicos y medir la dispersión de datos.



---
# <font color="red">1.3 Descripción estadística <p>
El análisis descriptivo resume las principales características de los datos mediante estadísticas fundamentales:

## <font color="blue">1.3.1 Medidas de tendencia central
* Media: Promedio de los datos
* Mediana: Valor central al ordenar los datos.
* Moda: Valor más frecuente en los datos.

## <font color="blue">1.3.2 Medidas de dispersión

* Varianza: Mide la variabilidad de los datos.
* Desviación estándar: Raíz cuadrada de la varianza, indica qué tan dispersos están los datos respecto a la media.
* Rango: Diferencia entre el valor máximo y mínimo.




---
# <font color="red">1.4 Análisis de distribuciones de datos <p>
## <font color="blue">1.4.1 Asimetría
Mide la simetría de la distribución.
* Asimetría positiva: Cola más larga a la derecha.
* Asimetría negativa: Cola más larga a la izquierda.

## <font color="blue">1.4.2 Curtosis:
Mide qué tan apuntada o plana es la distribución.
* Curtosis alta: Distribución con colas más pesadas (picos altos).
* Curtosis baja: Distribución más plana.

## <font color="blue">1.13.2 Mapa de calor para visualizar correlaciones:


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un dataset simulado (asegúrate de tener este paso antes si no lo tienes)
np.random.seed(42)
data = {
    "Estudiante": [f"E{i}" for i in range(1, 21)],
    "Edad": np.random.randint(18, 25, size=20),
    "Género": np.random.choice(["Femenino", "Masculino"], size=20),
    "Nota_Matemáticas": np.random.normal(70, 10, size=20).round(1),
    "Nota_Ciencias": np.random.normal(75, 12, size=20).round(1),
    "Nota_Inglés": np.random.normal(65, 15, size=20).round(1),
}
df = pd.DataFrame(data)

# Seleccionar solo las variables numéricas (int y float)
variables_numericas = df.select_dtypes(include=["int", "float"])

# Calcular la matriz de correlación
correlaciones = variables_numericas.corr()
# Create a mask to hide the lower triangle of the matrix
mask = np.triu(np.ones_like(correlaciones, dtype=bool))

# Crear el mapa de calor
sns.heatmap(correlaciones, annot=True, cmap="coolwarm", fmt=".2f", mask=mask)

# Añadir título al gráfico
plt.title("Mapa de Calor de Correlaciones")

# Mostrar el gráfico
plt.show()
```

# <font color="red">1.14 Ejemplo con apply para transformar datos
Crear una columna de aprobación en matemáticas:


```python
# Añadir columna indicando si aprobó matemáticas (nota >= 60)
df["Aprobado_Matemáticas"] = df["Nota_Matemáticas"].apply(lambda x: "Sí" if x >= 60 else "No")

df
```

```python
# Conteo de aprobaciones
aprobaciones = df["Aprobado_Matemáticas"].value_counts()

print("Conteo de aprobaciones en matemáticas:")
print(aprobaciones)
```

## <font color="blue">1.14.1 Visualización:


```python
# Gráfico circular para aprobaciones
aprobaciones.plot(kind="pie", autopct="%1.1f%%", colors=["lightgreen", "salmon"], startangle=90)
plt.title("Porcentaje de Aprobaciones en Matemáticas")
plt.ylabel("")  # Ocultar el label del eje Y
plt.show()
```
